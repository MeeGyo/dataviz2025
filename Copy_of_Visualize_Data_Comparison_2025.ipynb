```python
def swap_rows(tab, eth1='Asian', eth2='Other'):
    # Find the indices of 'Asian' and 'Other'
    asian_index = tab[tab['Officer-defined ethnicity'] == eth1].index[0]
    other_index = tab[tab['Officer-defined ethnicity'] == eth2].index[0]

    # Get the rows
    asian_row = tab.loc[asian_index].copy()
    other_row = tab.loc[other_index].copy()

    # Swap the rows
    tab.loc[asian_index] = other_row
    tab.loc[other_index] = asian_row

    return tab
```

```python
#Q Radar Chart



import numpy as np
import pandas as pd  #import pandas มาเพราะข้อมูลจะเกี่ยวกับตาราง
import matplotlib.pyplot as plt
from math import pi #import math มาเนื่องจากเกี่ยวกับวงกลมนำมาใช้เพื่อดึงค่าพาย

data_bed = pd.read_csv('/content/drive/MyDrive/Python&DataViz2025/2025-01/2025-01-bedfordshire-stop-and-search.csv')


data_london = pd.read_csv('/content/drive/MyDrive/Python&DataViz2025/2025-01/2025-01-city-of-london-stop-and-search.csv')

bed_ethnicity_counts = data_bed.groupby('Officer-defined ethnicity')['Type'].count().reset_index()
print("Bedfordshire Ethnicity Counts:")
display(bed_ethnicity_counts)

london_ethnicity_counts = data_london.groupby('Officer-defined ethnicity')['Type'].count().reset_index()
print("\nCity of London Ethnicity Counts:")
display(london_ethnicity_counts)

# Combine the two dataframes for easier plotting
combined_ethnicity_counts = pd.merge(bed_ethnicity_counts, london_ethnicity_counts, on='Officer-defined ethnicity', suffixes=('_Bedfordshire', '_London'))

categories = combined_ethnicity_counts['Officer-defined ethnicity'].tolist()
# categories += categories[:1] # Repeat the first category to close the circle - Removed this line

# Values for Bedfordshire and City of London
values_bed = combined_ethnicity_counts['Type_Bedfordshire'].tolist()
values_bed += values_bed[:1]

values_london = combined_ethnicity_counts['Type_London'].tolist()
values_london += values_london[:1]

# Calculate angles for each category
angles = [n / float(len(categories)) * 2 * pi for n in range(len(categories))] # Modified len(categories)-1 to len(categories)
angles += angles[:1]
############################################################################


# Plotting the radar chart
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6, 6), subplot_kw=dict(polar=True))

plt.xticks(angles[:-1], categories, color='grey', size=12)
ax.set_rlabel_position(0)

# Set the y-axis limits to a common maximum value
max_value = max(max(values_bed), max(values_london))
ax.set_ylim(0, max_value + 10) # Adding a small buffer for better visualization

# Plot Bedfordshire data
ax.plot(angles, values_bed, linewidth=2, linestyle='solid', label='Bedfordshire', color='#1f77b4')
ax.fill(angles, values_bed, '#1f77b4', alpha=0.25)

# Plot City of London data
ax.plot(angles, values_london, linewidth=2, linestyle='solid', label='City of London', color='#ff7f0e')
ax.fill(angles, values_london, '#ff7f0e', alpha=0.25)

plt.title('Comparison of Officer Ethnicity Counts between Bedfordshire and City of London', size=16, color='grey', y=1.1)
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
plt.show()
```

## Plot grouped bar chart ของ หนังอินเดีย โดย แยกกลุ่มตามประเภทหนัง และเปรียบเทียบ rating จาก ความยาวของหนัง

```python
data_india = pd.read_csv(os.path.join(path,'IMDb Movies India.csv'), encoding='latin-1')
data_india.head()
```

```python
data_india = data_india.dropna(subset=['Rating','Duration','Genre'])
data_india = data_india[['Rating','Duration','Genre']]
data_india
```

```python
# prompt: แปลงค่าใน column duration เป็น int โดยลบ min ออกและ เปลี่ยนชนิดข้อมูลเป็น integer

# Remove " min" from the 'Duration' column and convert to integer
data_india['Duration'] = data_india['Duration'].str.replace(' min', '').astype(int)
```

```python
data_india.describe()
```

```python
# prompt: แยกตารางออกเป็นตารางย่อย 5 ตารางสำหรับ genre: drama, action, romance, horror, musical

# Create a dictionary to store the tables
genre_tables = {}

# Define the genres to extract
genres = ['Drama', 'Action', 'Romance', 'Horror', 'Musical']

# Iterate through the genres and create separate tables
for genre in genres:
  genre_tables[genre] = data_india[data_india['Genre'] == genre].copy()

# Access the tables using the genre as the key
# Example:
# drama_table = genre_tables['Drama']
```

```python
all_table = pd.concat(genre_tables.values())
all_table
```

```python
# prompt: แบ่ง duration ออกเป็น short, medium, long โดย short สั้นกว่า 60 medium 60-90 ที่เหลือ เป็น long

# Function to categorize duration
def categorize_duration(duration):
  if duration < 60:
    return 'short'
  elif 60 <= duration <= 90:
    return 'medium'
  else:
    return 'long'

# Apply the function to create a new column
all_table['Duration Category'] = all_table['Duration'].apply(categorize_duration)
all_table
```

```python
all_table.groupby(['Genre','Duration Category'])['Rating'].mean()
```

## Radar Chart

```python
import numpy as np
import pandas as pd  #import pandas มาเพราะข้อมูลจะเกี่ยวกับตาราง
import matplotlib.pyplot as plt
from math import pi #import math มาเนื่องจากเกี่ยวกับวงกลมนำมาใช้เพื่อดึงค่าพาย
```

```python
data = {'client_id' : ['c1' , 'c2' , 'c3' , 'c4'],
         'cheese' : [2,5,4,3],
         'beer' : [2,1,3,5],
         'chicken' : [5,3,3,4],
        'bread' : [4,3,2,1],
       'coffee' : [2,3,5,3]}
df = pd.DataFrame.from_dict(data)
df
```

```python
categories = list(df.columns[1:])
categories
```

```python
values = df.iloc[:,1:].mean().values.flatten().tolist()
values += values[:1] # repeat the first value to close the circular graph
angles = [n / float(len(categories)) * 2 * pi for n in range(len(categories))]
angles += angles[:1]
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 8),
                       subplot_kw=dict(polar=True))

plt.xticks(angles[:-1], categories, color='grey', size=12)
plt.yticks(np.arange(1, 6), ['1', '2', '3', '4', '5'],
           color='grey', size=12)

plt.ylim(0, 5)
ax.set_rlabel_position(30)

ax.plot(angles, values, linewidth=1, linestyle='solid' , color = '#ea3033') #เส้นกราฟ
ax.fill(angles, values, 'lightpink', alpha=0.5) #พื้นที่
```

```python
angles = [n / float(len(categories)) * 2 * pi for n in range(len(categories))]
angles += angles[:1]

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 8),
                       subplot_kw=dict(polar=True))

plt.xticks(angles[:-1], categories, color='grey', size=12)
plt.yticks(np.arange(1, 6), ['1', '2', '3', '4', '5'],
           color='grey', size=12)

plt.ylim(0, 5)
ax.set_rlabel_position(30)

# part 1
val_c1 = df.loc[0].drop('client_id').values.flatten().tolist()
val_c1 += val_c1[:1]
ax.plot(angles, val_c1, linewidth=1,
        linestyle='solid', label='Client c1')
ax.fill(angles, val_c1, 'skyblue', alpha=0.4)

# part 2
val_c2=df.loc[1].drop('client_id').values.flatten().tolist()
val_c2 += val_c2[:1]
ax.plot(angles, val_c2, linewidth=1,
        linestyle='solid', label='Client c2')
ax.fill(angles, val_c2, 'lightpink', alpha=0.4)

plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
```
